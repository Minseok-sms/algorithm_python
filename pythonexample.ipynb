```python
class Person:
    def __init__(self,name,age): #constructor 랑비슷
        self.name = name;
        self.age = age;
         
    def say_hello(self, to_name):
        print("안녕! " + to_name + " 나는 " + self.name );
    def introduce(self):
        print("내이름은 " + self.name + str(self.age));


w = Person("민석",20);
w.introduce()

class Police(Person): #상속
    def arrest(self, to_arrest):
        print("넌 채포됐다, " + to_arrest);
class Programmer(Person):
    def program(self, to_program):
        print("다음엔 뭘만들지 ?" + to_program );
mins = Person("민석",20);
jenny = Police("제니",20);
michale = Police("마이클",22);

jenny.introduce();
jenny.arrest("안뇽");

```
